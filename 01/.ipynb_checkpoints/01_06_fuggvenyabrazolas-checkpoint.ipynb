```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)*(1-4*np.exp(0.7*x)/(x**3))
```

```python
x1 = np.linspace(2,10,30)
y1 = [f(x) for x in x1]
```

```python
fig1 = plt.figure(num = 1, figsize=(16/2.54,10/2.54))
ax = fig1.add_subplot(111)
ax.plot(x1,y1)

# tengelyfeliratok
plt.xlabel('x')
plt.ylabel('f(x)')
# rácsozás
ax.grid()

# ábra mentés
plt.savefig(".\\abra1.png", bbox_inches="tight")
plt.savefig(".\\abra1.pdf", bbox_inches="tight")

plt.show()
```

```python
def g(x):
    return np.sqrt(x)
def h(x):
    return x**2
```

```python
x2 = np.linspace(0,1,50)
y2 = [g(x) for x in x2]
y3 = [h(x) for x in x2]
```

Ha több függvényt ábrázolunk akkor automatikusan jelöli őket:

```python
fig2 = plt.figure(num = 2, figsize=(16/2.54,10/2.54))
ax = fig2.add_subplot(111)
ax.plot(x2,y2)
ax.plot(x2,y3)

plt.xlabel('x')
# felirat
ax.legend((r'$\sqrt{x}$',r'$x^2$'), loc = 1)
ax.grid()

plt.savefig(".\\abra2.pdf", bbox_inches="tight")

plt.show()
```

```python
fig3 = plt.figure(num = 3, figsize=(16/2.54,10/2.54))
ax = fig3.add_subplot(111)
ax.plot(x1,y1)

plt.xlabel('x')
plt.ylabel('f(x)')
# megjelenítési határ
plt.ylim(bottom = -0.5, top = 0.5)
ax.grid()

plt.show()
```

```python
x3 = np.arange(1,14+1)
y3 = [np.exp(x)/(1.4e4) for x in x3]
fig4 = plt.figure(num = 4, figsize=(16/2.54,10/2.54) )
ax = fig4.add_subplot(111)
ax.plot(x3, y3, linestyle = '', marker = 'o')
plt.xlabel('szorgalmi hét sorszáma')
plt.ylabel(r'hallgatók terheltsége [%]-ban')
ax.grid()
plt.show()
```

```python
fig5 = plt.figure(num = 5, figsize=(16/2.54,10/2.54) )
ax = fig5.add_subplot(111)
ax.plot(x3, y3, linestyle = '', marker = 'o')
plt.yscale('log')
plt.xlabel('szorgalmi hét sorszáma')
plt.ylabel(r'hallgatók terheltsége [%]-ban')
ax.grid()
plt.show()
```

```python
fig6 = plt.figure(num = 6, figsize=(16/2.54,10/2.54) )
ax = fig6.add_subplot(111)
ax.plot(x3, y3, linestyle = '', marker = 'o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('szorgalmi hét sorszáma')
plt.ylabel(r'hallgatók terheltsége [%]-ban')
ax.grid()
plt.show()
```

Parametrikus függvényábrázolás:

```python
t = np.linspace(0,2*np.pi,100)
x = [np.sin(t) for t in t]
y = [np.cos(t) for t in t]
```

```python
fig7 = plt.figure(num = 7, figsize=(16/2.54,10/2.54) )
ax = fig7.add_subplot(111)
ax.plot(x, y, linestyle = '--')
ax.grid()
plt.axis('equal')
plt.show()
```

3D függvényábrázolás:

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x, y):
    return np.sin(np.exp( - x ** 2 - y ** 2))

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure(figsize=(16/2.54,10/2.54))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```
